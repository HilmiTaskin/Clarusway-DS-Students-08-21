## 1b: Swarm Plot

## 2: Plot 12 most dangerous cities

```python
df.dropna(inplace=True)
df.city.value_counts(dropna=False)
```

```python
cities = df.city.value_counts().index[:13]
counts = df.city.value_counts().values[:13]
```

```python
fig,ax = plt.subplots(figsize=(12,8))
sns.barplot(data=df, x=cities, y=counts, ax=ax)
plt.xticks(rotation=70);
```

```python
# with seaborn barplot

```

```python
dfc = df.city.value_counts(dropna=False)[:13]
dfc.index
```

```python
order=pd.value_counts(df.city).iloc[:13].index
order
```

```python
fig,ax = plt.subplots(figsize=(12,8))
sns.countplot(data=df, x=df.city, ax=ax, order=pd.value_counts(df.city).iloc[:13].index)
plt.xticks(rotation=70);
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states
