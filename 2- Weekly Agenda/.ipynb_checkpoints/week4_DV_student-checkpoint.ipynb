# 2. Make a bar plot

```python
labels = ['A', 'B', 'C','D', 'E', 'F', 'G']
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, income, width, label='income')
rects2 = ax.bar(x + width/2, expense, width, label='expense')

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

```

# 3. Rotate the labels and set figsize

```python
labels = ['A', 'B', 'C','D', 'E', 'F', 'G']
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
ind = [x for x, _ in enumerate(labels)]
plt.figure(figsize=(10,6))
plt.bar(ind, income, width=0.4, label='expense')
plt.bar(ind, expense, width=0.4, label='income',bottom=income)


plt.xticks(ind, labels)
plt.legend(loc="upper right")

```

# 4. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"
# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
labels = ["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(7.5,6))
ax.bar(x - width/2, income, width, label='income')
ax.bar(x + width/2, expense, width, label='expense')

   
for p in ax.patches:
    ax.annotate(str(p.get_height()),(p.get_x()+0.1, p.get_height()+2))
    
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()
```

# 9. Make a scatter plot between tip and total_bill

```python
fig,ax = plt.subplots()
ax.scatter(data=tips,x="tip",y="total_bill",s=20)
ax.set_xlabel("tip")
ax.set_ylabel("total_bill");
```

# 10. Set an additional dimension using size column

```python
plt.figure(figsize=(12,8))
plt.scatter(data=tips,x="tip",y="total_bill",c="size",cmap="coolwarm",s=20)
ax.set_xlabel("tip")
ax.set_ylabel("total_bill")
plt.colorbar();
```
